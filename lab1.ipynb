### Polazni zadatak

Kompozicija noćnog vlaka ima barem jedna spavaća kola, barem dvoja kola prvog razreda, dok su ostali vagoni drugog razreda. Spavaća kola primaju 30 putnika, vagon prvog razreda 60, a vagoni drugog razreda po 90. Mora se složiti kompozicija za barem 600 putnika. Trošak spavaćih kola je 400kn na noć, vagon prvog razreda košta 250kn na noć, dok vagon drugog razreda košta 100kn po jednoj noći. Kako sastaviti kompoziciju koja može primiti očekivani broj putnika, a da ima najmanje troškove?

### Matematički model zadanog problema

**Definiramo varijable:**
- 1. broj spavaćih kola: $$x_1$$ 
- 2. broj vagona prvog razreda: $$x_2$$
- 3. broj vagona drugog razreda: $$x_3$$

**Ciljna funkcija:**  
Minimizacija troškova kompozicije vagona:
$$
\text{Min } Z = 400x_1 + 250x_2 + 100x_3
$$

**Ograničenja:**

1. Broj spavaćih kola mora biti barem 1:
$$
x_1 \geq 1
$$

2. Broj vagona prvog razreda mora biti barem 2:
$$
x_2 \geq 2
$$

3. Ukupni kapacitet mora biti najmanje 600 putnika:
$$
30x_1 + 60x_2 + 90x_3 \geq 600
$$

4. Ne možemo imati negativan broj vagona:
$$
x_1 \geq 0, \quad x_2 \geq 0, \quad x_3 \geq 0
$$


### Matematički model dualnog problema

**Definiramo dualne varijable:**
- \(y_1\): vrijednost povezana s ograničenjem kapaciteta $$30x_1 + 60x_2 + 90x_3 \geq 600$$  
- \(y_2\): vrijednost povezana s minimalnim brojem spavaćih kola $$x_1 \geq 1$$ 
- \(y_3\): vrijednost povezana s minimalnim brojem vagona prvog razreda $$x_2 \geq 2$$  

**Ciljna funkcija:**  
Maksimizacija ukupne vrijednosti resursa:
$$
\text{Max } W = 600y_1 + y_2 + 2y_3
$$

**Ograničenja:**

1. Ograničenje za trošak spavaćih kola (\(x_1\)):  
   $$ 
   30y_1 + y_2 \leq 400 
   $$

2. Ograničenje za trošak vagona prvog razreda (\(x_2\)):  
   $$ 
   60y_1 + y_3 \leq 250 
   $$

3. Ograničenje za trošak vagona drugog razreda (\(x_3\)):  
   $$ 
   90y_1 \leq 100 
   $$

4. Nenegativnost dualnih varijabli:
   $$
   y_1 \geq 0, \quad y_2 \geq 0, \quad y_3 \geq 0
   $$


### Softversko i grafičko rješenje

Softversko numeričko rješenje postignuto je korištenjem biblioteke `scipy` u programskom jeziku Python, koja u sebi sadrži primalni i dualni model.

Postoji jedno optimalno rješenje funkcije cilja i ono iznosi:
$$
Z = W = 1400 \, \text{kn troška}.
$$

**Primalne varijable:**
- Spavaća kola (\(x_1\)): \(1\) vagon  
- Kola prve klase (\(x_2\)): \(2\) vagona  
- Kola druge klase (\(x_3\)): \(5\) vagona  

**Dualne varijable:**
- \(y_1 = 1.11\)  
- \(y_2 = 366.67\)  
- \(y_3 = 183.33\)  


### Iskorištenost resursa

Resursi su u potpunosti iskorišteni jer su broj putnika od \(600\) i količina svakog pojedinog vagona u potpunosti iskorišteni da bi se dostigao optimum. Nema slobodnog prostora za neko optimalnije rješenje.

**Ograničenje broja putnika:**
$$
30x_1 + 60x_2 + 90x_3 \geq 600
$$
Izračun:
$$
30 \cdot 1 + 60 \cdot 2 + 90 \cdot 5 \geq 600
$$
$$
600 \geq 600
$$

**Ograničenje broja vagona:**
1. Za spavaća kola (\(x_1\)):  
   $$
   x_1 \geq 1
   $$
   Izračun:
   $$
   1 \geq 1
   $$

2. Za vagone prve klase (\(x_2\)):  
   $$
   x_2 \geq 2
   $$
   Izračun:
   $$
   2 \geq 2
   $$


### Analiza osjetljivosti

Za analizu osjetljivosti koristimo sjenu cijene (shadow price) kako bismo odredili utjecaj promjene pojedinog ograničenja na ciljnu funkciju. Sjena cijene izračunava se pomoću formule:


$$
SP = \frac{C_B - C_A}{Z_B - Z_A}
$$



Gdje su:
- \( Z_B \): vrijednost ciljne funkcije nakon promjene ograničenja,
- \( Z_A \): vrijednost ciljne funkcije prije promjene ograničenja,
- \( C_B \): vrijednost resursa nakon promjene ograničenja,
- \( C_A \): vrijednost resursa prije promjene ograničenja.



### Shadow Price za kapacitet $$30x_1 + 60x_2 + 90x_3 \geq 600$$

#### Formula:
$$
SP = \frac{Z_B - Z_A}{C_B - C_A}
$$

#### Vrijednosti:
- Početna vrijednost resursa (\(C_A = 600\)),
- Promijenjena vrijednost resursa (\(C_B = 601\)),
- Početna vrijednost ciljne funkcije (\(Z_A = 1400\)),
- Promijenjena vrijednost ciljne funkcije (\(Z_B = 1401.11\)).

#### Izračun:
$$
SP = \frac{1401.11 - 1250}{601 - 600}
$$
$$
SP = \frac{1.11}{1} = 1.11
$$

---

### Shadow Price za minimalni broj spavaćih kola $$x_1 \geq 1$$

#### Vrijednosti:
- Početna vrijednost resursa (\(C_A = 1\)),
- Promijenjena vrijednost resursa (\(C_B = 2\)),
- Početna vrijednost ciljne funkcije (\(Z_A = 1400\)),
- Promijenjena vrijednost ciljne funkcije (\(Z_B = 1766.67\)).

#### Izračun:
$$
SP = \frac{1766.67 - 1400}{2 - 1}
$$
$$
SP = \frac{366.67}{1} = 366.67
$$

---

### Shadow Price za minimalni broj vagona prvog razreda $$x_2 \geq 2$$

#### Vrijednosti:
- Početna vrijednost resursa (\(C_A = 2\)),
- Promijenjena vrijednost resursa (\(C_B = 3\)),
- Početna vrijednost ciljne funkcije (\(Z_A = 1400\)),
- Promijenjena vrijednost ciljne funkcije (\(Z_B = 1583.33\)).

#### Izračun:
$$
SP = \frac{1583.33 - 1400}{3 - 2}
$$
$$
SP = \frac{183.33}{1} = 183.33
$$

---

### Rezultat:

1. **Shadow Price za kapacitet:** \( SP = 1.11 \)
2. **Shadow Price za spavaća kola:** \( SP = 366.67 \)
3. **Shadow Price za vagone prvog razreda:** \( SP = 183.33 \)



### Cijeli brojevi

Zadatak je bilo potrebno rješavati kao cjelobrojni problem linearnog programiranja zato što broj vagona ili broj putnika ne može biti decimalni broj.

```python
#potrebni importovi

#from gilp.simplex import LP
#from gilp.visualize import lp_visual
from scipy.optimize import linprog

import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D
```

```python
#primalno - programski

# Parametri problema - koeficijenti za minimizaciju troskova
c = [400, 250, 100]  #trosak spavacih kola 400kn/noc, prvog razreda 250kn/noc, drugog razreda 100kn/noc
A = [[-30, -60, -90]]  #spavaca kola primaju 30 putnika, prva klasa 60, a druga 90 putnika, negativni brojevi zbog linprog gdje je uvjet da lijeva strana bude <= desnoj strani
b = [-600]  #ukupni kapacitet je barem 600 putnika
bounds = [(1, None), (2, None), (0, None)]  #barem jedna spavaća kola, dvoja kola 1. razreda, ostali mogu biti 0 ili više


# Rješavanje problema linearne optimizacije
result = linprog(c=c, 
                 A_ub=A, 
                 b_ub=b, 
                 bounds=bounds, 
                 method='highs')


# Ispis rezultata
if result.success:
    optimal_x, optimal_y, optimal_z = result.x
    print(f'Optimalna kompozicija:')
    print(f'Spavaća kola: {int(optimal_x)} vagon')
    print(f'Kola prvog razreda: {int(optimal_y)} vagona')
    print(f'Kola drugog razreda: {int(optimal_z)} vagona')
    print(f'Minimalni trošak: {int(result.fun)} kn')
else:
    print('Rješenje nije izvedivo')
```

```python
#dualno - programski

# Koeficijenti ciljne funkcije (negirani zbog maksimizacije)
c = [-600, -1, -2]  # Maksimizacija

# Matrica ograničenja (koeficijenti)
A = [
    [30, 1, 0],   # Spavaća kola
    [60, 0, 1],   # Vagon prvog razreda
    [90, 0, 0]    # Vagon drugog razreda
]

# Desna strana ograničenja (troškovi)
b = [400, 250, 100]

# Granice za varijable (sve su nenegativne)
bounds = [(0, None), (0, None), (0, None)]

# Rješavanje problema
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Ispis rezultata
if result.success:
    print("Optimalno rješenje dualnog problema:")
    print(f"y1: {result.x[0]:.2f}")
    print(f"y2: {result.x[1]:.2f}")
    print(f"y3: {result.x[2]:.2f}")
    print(f"Maksimalna vrijednost (W): {-result.fun:.2f}")
else:
    print("Problem nije riješen.")
```

```python
#graficko rjesenje

x1 = np.arange(1, 10, 1)  
x2 = np.arange(2, 10, 1)  
x3 = np.arange(0, 10, 1)  

X1, X2, X3 = np.meshgrid(x1, x2, x3)

passenger_constraint = 30 * X1 + 60 * X2 + 90 * X3 >= 600

Z = 400 * X1 + 250 * X2 + 100 * X3

feasible_points = passenger_constraint & (X1 >= 1) & (X2 >= 2)

A = np.array([X1[feasible_points], X2[feasible_points], X3[feasible_points]])
C = np.array([Z[feasible_points]])

xx, yy = np.meshgrid(np.linspace(1, 10, 10), np.linspace(2, 10, 10))
zz = (600 - 30 * xx - 60 * yy) / 90
zz[zz < 0] = 0  

optimal_index = np.argmin(C[0])
optimal_point = np.array([1, 2, 5])
optimal_cost = 400 * optimal_point[0] + 250 * optimal_point[1] + 100 * optimal_point[2]

fig = plt.figure(figsize=(8, 6))
colors = np.where((A[0] == optimal_point[0]) & (A[1] == optimal_point[1]) & (A[2] == optimal_point[2]), 'red', 'blue')  # Correctly mark the optimal point
ax = fig.add_subplot(111, projection='3d')

ax.scatter(A[0], A[1], A[2], c=colors, label="Feasible Points")

ax.scatter(optimal_point[0], optimal_point[1], optimal_point[2], c='red', s=60, edgecolor='black', label=f"Optimal Point: {optimal_point}, Cost: {optimal_cost} kn")

ax.plot_surface(xx, yy, zz, color='grey', alpha=0.5)

ax.set_xlabel("$x_1$ (Spavaća kola)", labelpad=10)
ax.set_ylabel("$x_2$ (Vagoni 1. razreda)", labelpad=10)
ax.set_zlabel("$x_3$ (Vagoni 2. razreda)", labelpad=10)
ax.set_title("Feasible Region and Optimal Solution", pad=20)

custom_lines = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label=f"Optimal Point: {optimal_point}, Cost: {optimal_cost} kn"),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Feasible Points'),
    plt.Line2D([0], [0], color='grey', lw=2, alpha=0.5, label='Passenger Constraint')
]
ax.legend(handles=custom_lines, loc='upper right')

plt.tight_layout()
plt.show()
```
